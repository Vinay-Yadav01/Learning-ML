```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Melbourne_housing_FULL.csv')
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
col_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea',
              'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[col_to_use]
df.head()
```

```python
df.isna().sum()
```

```python
zero_fill = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[zero_fill] = df[zero_fill].fillna(0)
df.isna().sum()
```

```python
mean_fill = ['Landsize', 'BuildingArea']
df[mean_fill] = df[mean_fill].fillna(df[mean_fill].mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)  # Get dummies
boolean_cols = [col for col in df if df[col].dtype == 'bool']  # Find boolean columns
df[boolean_cols] = df[boolean_cols].astype(int)  # Convert to int
df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```
