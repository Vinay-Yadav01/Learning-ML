```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('revenue.csv')
df
```

```python
plt.bar(df['company'], df['revenue'])
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Revenue of Companies')
plt.show()
```

```python
plt.bar(df['company'], df['revenue'])
plt.yscale('log')  # Set the y-axis to logarithmic scale
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Revenue of Companies')
plt.show()
```

```python
df = pd.read_csv('income.csv')
df.head()
```

```python
import seaborn as sns
sns.barplot(x='income($)', y='count', data=df)
plt.xticks(rotation=70, ha='right')
plt.show()
```

```python
import math
df['log income($) count'] = df['count'].apply(lambda x: math.log(x))
df.head()
```

```python
sns.barplot(x='log income($) count', y='count', data=df)
plt.xticks(rotation=70, ha='right')
plt.show()
```
