```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('revenue.csv')
df
```

```python
plt.bar(df['company'], df['revenue'])
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Revenue of Companies')
plt.show()
```

```python
plt.bar(df['company'], df['revenue'])
plt.yscale('log')  # Set the y-axis to logarithmic scale
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Revenue of Companies')
plt.show()
```
