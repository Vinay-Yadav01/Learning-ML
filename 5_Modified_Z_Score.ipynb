```python
import pandas as pd
```

```python
df = pd.read_csv('/content/movie_revenues.csv')
df.head()
```

```python
df['revenue_mln'] = df['revenue'].apply(lambda x: x/1000000)
df.revenue_mln.describe()
```

```python
_, mean, std, *_ = df.revenue_mln.describe()
```

```python
mean,std
```

```python
def get_z_score(value,mean,std):
  return (value-mean)/std
```

```python
df['zscore'] = df.revenue_mln.apply(lambda x: get_z_score(x,mean,std))
df.head()
```

```python
df[(df.zscore>3) |
   (df.zscore<-3)]
```

```python
import numpy as np
def get_mad(s):
  median = np.median(s)
  diff = abs(s-median)
  MAD = np.median(diff)
  return MAD
```

```python
MAD = get_mad(df.revenue_mln)
median = np.median(df.revenue_mln)
MAD, median
```

```python
def get_modified_zscore(x, median, MAD):
  return 0.6745*(x-median)/MAD
```

```python
get_modified_zscore(2787,median,MAD)
```

```python
df['modified_zscore'] = df.revenue_mln.apply(lambda x: get_modified_zscore(x,median,MAD))
df.head()
```

```python
df[df.modified_zscore>3.5]
```
